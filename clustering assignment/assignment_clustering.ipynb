2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
# Question 1 part 1
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
# Question 1 part 2
import matplotlib.pyplot as plt
import seaborn as sns

# Store datasets in a dictionary
datasets = {
    "Noise 0.125": df0_125,
    "Noise 0.25": df0_25,
    "Noise 0.5": df0_5,
    "Noise 1.0": df1_0,
    "Noise 2.0": df2_0,
}

# Generate scatterplots for each dataset with color-coded groups
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for ax, (title, df) in zip(axes, datasets.items()):
    sns.scatterplot(x=df["x1"], y=df["x2"], hue=df["group"], ax=ax, palette="Set1", alpha=0.8)
    ax.set_title(title)
    ax.legend(title="Group")

plt.suptitle("Scatterplots of (X1, X2) by Group for Different Noise Levels", fontsize=14)
plt.show()

# The clusters are more mixed with eeach other compared to lower noise situations.
````

````python
# Question 1 part 3
from sklearn.cluster import KMeans

# Function to compute inertia for different values of k
def compute_inertia(df, max_k=10):
    X = df[['x1', 'x2']].values
    inertia = []
    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=100, n_init=10)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)
    return inertia

# Compute inertia for each dataset
inertia_results = {title: compute_inertia(df) for title, df in datasets.items()}

# Plot scree plot (Elbow Method) for all datasets in a single figure
fig, ax = plt.subplots(figsize=(8, 5))

for title, inertia in inertia_results.items():
    ax.plot(range(1, 11), inertia, marker='o', label=title)

ax.set_xlabel("Number of Clusters (k)")
ax.set_ylabel("Inertia")
ax.set_title("Scree Plot (Elbow Method) for Different Noise Levels")
ax.legend()
plt.show()

#As the level of noise increases in the datasets, the clarity of the elbow in the scree plot diminishes, making it more difficult to definitively select an optimal k.
#For low noise levels (0.125, 0.25, 0.5), the scree plot exhibits a clear elbow around k=3.
````

# Question 1 part 4

The elbow method is used to determine the optimal number of clusters (
k). The core idea is to find an "elbow" in the scree plot, which represents the relationship between the number of clusters and the inertia (within-cluster variance). Inertia measures how compact the clusters are, with lower values indicating better clustering. The goal is to select the value of
k where adding more clusters no longer significantly reduces inertia, meaning the clustering no longer improves substantially.

In the given scree plot, we observe that for low noise levels (0.125, 0.25, 0.5), there is a clear elbow around

=
3
. This aligns with the original dataset, which was generated with three underlying clusters. At this point, inertia drops significantly before flattening out, indicating that beyond

=
3
, adding more clusters provides diminishing returns in terms of reducing within-cluster variance. The presence of a distinct elbow means the clustering structure is well-defined, making it easier to determine the appropriate number of clusters.

However, as the noise level increases (1.0, 2.0), the elbow becomes less pronounced. The inertia curve smoothly declines instead of sharply bending, making it difficult to pinpoint a single best
. This suggests that the data is becoming more dispersed and overlapping, reducing the distinctiveness of clusters. As a result, clustering becomes more ambiguous, and additional validation methods such as the silhouette score or gap statistic may be necessary to confirm the best choice of

**Q4.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
#question 4 part 1
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Load Airbnb dataset
url = "https://raw.githubusercontent.com/ds4e/clustering/refs/heads/main/assignment/data/airbnb_hw.csv"
df = pd.read_csv(url,low_memory=False)
print(df)

df = df.dropna(subset=['Price'])  # Remove rows where Price is NaN

#print(df.head())
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

print(df.head())

# Select relevant features for clustering
df = df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']]

# Remove missing values (K-Means does not handle NaN values)
df = df.dropna()

````

````python
# Question 4 part 2
from sklearn.preprocessing import MinMaxScaler

# Define Min-Max normalization function
def maxmin(x):
    return (x - x.min()) / (x.max() - x.min())

# Apply Min-Max normalization to the numerical columns
df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']] = df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']].apply(maxmin)

# Display the first few rows to check the data
print(df.head())
````

````python
# Question 4 part 3
from sklearn.cluster import KMeans

# Select features for clustering
X = df[['Beds', 'Number Of Reviews', 'Review Scores Rating']]

# Apply K-Means clustering with k=6
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)

# Display the mean values of each cluster
print(df.groupby('Cluster').mean())
````

````python
# Question 4 part 4
sns.pairplot(df, hue='Cluster', diag_kind='kde')
plt.show()
````

````python
# Question 4 part 5
# Compute summary statistics for each cluster
cluster_summary = df.groupby('Cluster')['Price'].describe()
print(cluster_summary)

# Identify the cluster with the highest average rental price
highest_price_cluster = cluster_summary['mean'].idxmax()
````

````python
sse = []  # List to store sum of squared errors
k_range = range(1, 11)  # Test k values from 1 to 10

# Compute SSE for different k values
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # Store SSE (inertia)

# Plot Scree Plot (Elbow Method)
plt.figure(figsize=(6, 4))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Scree Plot for Optimal k')
plt.show()
````
